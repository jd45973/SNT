Les langages du web Plusieurs langages ont été standardisés par le W3C pour décrire le contenu des pages (html), leur forme (css) et pour programmer les interactions (JavaScript).

Le format HTML est un format textuel permettant de décrire le contenu et la structure d'un document (ou une page) Web.

## Langage HTML


Le HTML et en particulier HTML5 est un langage à balises. 
Une page en HTML est composée :
* Une entête (head) : elle contient les informations concernant le fichier, comme les métadonnées, le lien vers le fichier CSS, le titre de la page …
* Un corps (body) : composé d’une entête de page (header), menu, sections …
* Un pied de page (footer) : il contient le statut du blog voire du site, contact …  
Grâce un éditeur de texte, on crée le contenu de notre page WEB, en utilisant le langage HTML. La mise en forme de notre page WEB est réalisée en utilisant le CSS.


### Norme HTML et validation

Le format HTML est standardisé par le W3C. Ce dernier propose un validateur en ligne qui permet de vérifier si un fichier HTML est conforme au standard. Il est accessible à l'adresse :  
https://validator.w3.org

L'élément html doit contenir un attribut **lang**, cette information peut être utilisée par les navigateurs pour proposer de la correction orthographique par exemple.

#### Une entête (head) 
elle contient les informations concernant le fichier, comme les métadonnées, le lien vers le fichier CSS, le titre de la page …
L'entête contient au moins un élément *title*  
La balise vide **&lt;meta charset="utf-8 /&gt;** indique l'encodage de caractères utilisé par le document (il devrait toujours valoir utf-8, les autres encodages étant tombés en désuétude)

L'élément body représente le corps et regroupe tous les éléments qui structurent la page. Ces derniers sont affichés dans la zone d'affichage du navigateur Web.

Ainsi on les écrira **&lt;br /&gt;** par exemple pour la balise retour à la ligne.

### Liens et ressources

La fonctionnalité principale des documents HTML est d'utiliser des liens *hypertextes*.

Ces derniers sont insérés grâce à la balise **&lt;a&gt;**.

Cette dernière se découpe en deux parties : La première, un attribut ***href*** donnant l'emplacement de la ressource, la seconde est le contenu de la balise qui sert à l'affichage du lien.


<!DOCTYPE html>
<html lang="fr">
<head>
    <title> Un titre </title>
    <meta charset="utf-8">
</head>
<body>
    <a href="https://littre.etab.ac-caen.fr/"> Le lien hypertexte pour le <b>site du lycée</b> </a>
</body>
</html>

Pour insérer des images on utilise la balise **&lt;img /&gt;**. Cette balise doit contenir deux attributs, **src** qui indique l'emplacement de l'image et **alt** qui donne un texte alternatif au cas ou le l'image ne se charge pas. Pour tester le "texte alternatif" vous pouvez changer le nom du fichier image dans le code suivant.

<!DOCTYPE html>
<html lang="fr">
<head>
    <title> Un titre </title>
    <meta charset="utf-8">
</head>
<body>
    <img src="lycee.jpg"" alt="image du lycée" />
</body>
</html>

Un fichier CSS (de l’anglais Cascading Stylesheet) contient des règles de mise en forme. Par exemple, tous les titres de niveau 1 (marqués par les balise **&lt;h1&gt;**) sont de telle couleur et de telle police.

## Evènements

Nous allons voir très sommairement comment une page Web peut  avec l'utilisateur via des clics, des réactios au survol de la souris, effectuer des apparitions de fenêtres... LE langage JavaScript permet de mettre en place ce qe l'on nomme des **évènements**.

## Les types d'évènements les plus communs

|Evènement|Description|
|:---------|:---------|
|onchange |Un élément HTML a été modifié.|
|onclick  |L'utilisateur a cliqué sur un élément HTML|
|onmouseover|L'utilisateur a survolé avec la souris au-dessus d'un élément HTML.|
|onmouseout|L'utilisateur a enlevé la souris qui était au-dessus d'un élément HTML.|
|onkeydown|L'utilisateru aa en foncé une touche clavier.|
|onload|La page a fini d'être chargée.|
